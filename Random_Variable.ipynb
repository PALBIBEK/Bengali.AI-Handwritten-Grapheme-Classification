```python
from scipy import stats

data = [4.8, 4.9, 5.1, 5.2, 5.3, 5.4, 5.5, 5.6, 5.7]

# Perform Shapiro-Wilk test
# stat reprsents the value of W in Shapiro-Wilk test which basically ranges between 0 to 1. Higher the value of W, it is more likely that the data or
# sample follows normal distribution

stat, p = stats.shapiro(data)

print('Statistics=%.3f, p=%.3f' % (stat, p))

# Set the significance level
alpha = 0.05

# Decision rule
if p > alpha:
    print("Sample looks Gaussian (fail to reject H0)")
else:
    print("Sample does not look Gaussian (reject H0)")
```

```python
import numpy as np
from scipy.stats import ks_1samp, ks_2samp, norm

# One-Sample K-S Test
# Sample data
sample_data = [0.1, -0.2, 0.3, -1.1, 0.5, 0.0, -0.5, 1.2, -0.3, 0.4]

# Perform the one-sample K-S test against the standard normal distribution
ks_statistic, p_value = ks_1samp(sample_data, norm.cdf)

print("One-Sample K-S Test")
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value > alpha:
    print("Fail to reject the null hypothesis: The sample comes from the reference distribution.")
else:
    print("Reject the null hypothesis: The sample does not come from the reference distribution.")

# Two-Sample K-S Test
# Sample data
sample_data1 = [1.1, 2.3, 3.2, 4.8, 5.5]
sample_data2 = [1.0, 2.5, 3.3, 4.7, 5.6]

# Perform the two-sample K-S test
ks_statistic, p_value = ks_2samp(sample_data1, sample_data2)

print("\nTwo-Sample K-S Test")
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")

# Interpretation
if p_value > alpha:
    print("Fail to reject the null hypothesis: The two samples come from the same distribution.")
else:
    print("Reject the null hypothesis: The two samples come from different distributions.")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import ks_1samp, ks_2samp, norm
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris.head())

# Extract sepal length for two species
setosa_sepal_length = iris[iris['species'] == 'setosa']['sepal_length']
versicolor_sepal_length = iris[iris['species'] == 'versicolor']['sepal_length']

# Print a brief summary
print("Setosa Sepal Length:\n", setosa_sepal_length.describe())
print("Versicolor Sepal Length:\n", versicolor_sepal_length.describe())

# Perform the one-sample K-S test for Setosa sepal length against the standard normal distribution
ks_statistic, p_value = ks_1samp(setosa_sepal_length, norm.cdf, args=(setosa_sepal_length.mean(), setosa_sepal_length.std()))

print("One-Sample K-S Test for Setosa Sepal Length")
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value > alpha:
    print("Fail to reject the null hypothesis: The sample comes from the reference distribution.")
else:
    print("Reject the null hypothesis: The sample does not come from the reference distribution.")

# Perform the two-sample K-S test for sepal length between Setosa and Versicolor
ks_statistic, p_value = ks_2samp(setosa_sepal_length, versicolor_sepal_length)

print("\nTwo-Sample K-S Test for Sepal Length between Setosa and Versicolor")
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")

# Interpretation
if p_value > alpha:
    print("Fail to reject the null hypothesis: The two samples come from the same distribution.")
else:
    print("Reject the null hypothesis: The two samples come from different distributions.")
```

```python
import random
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

class WeightedDatasetSplitter:
    def __init__(self, dataset, feature_name, target_name):
        self.dataset = dataset
        self.feature_name = feature_name
        self.target_name = target_name
        self.df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
        self.df['species'] = dataset.target
        self.target_names = dataset.target_names
        self.average_values = self.calculate_average_values()

    def calculate_average_values(self):
        feature_values = self.df[self.feature_name]
        average_values = [
            np.mean(feature_values[self.df[self.target_name] == i])
            for i in range(len(self.target_names))
        ]
        return average_values

    def weighted_random_pick(self, items, values):
        """
        Picks an item from the items list with a probability proportional to its corresponding value in the values list.
        """
        if len(items) != len(values):
            raise ValueError("Items and values must have the same length")

        # Calculate the total sum of all values
        total = sum(values)

        # Generate a list of probabilities proportional to values
        probabilities = [value / total for value in values]

        # Use random.choices to pick an item based on the given probabilities
        picked_item = random.choices(items, weights=probabilities, k=1)

        return picked_item[0]

    def split_dataset(self):
        train_list = []
        test_list = []

        for i in range(len(self.df)):
            picked_species = self.weighted_random_pick(
                list(range(len(self.target_names))),
                self.average_values
            )
            if self.df.iloc[i]['species'] == picked_species:
                train_list.append(self.df.iloc[i])
            else:
                test_list.append(self.df.iloc[i])

        train_df = pd.DataFrame(train_list, columns=self.df.columns)
        test_df = pd.DataFrame(test_list, columns=self.df.columns)

        return train_df, test_df

# Example usage
iris = load_iris()
splitter = WeightedDatasetSplitter(
    dataset=iris,
    feature_name='sepal length (cm)',
    target_name='species'
)

train_df, test_df = splitter.split_dataset()
print("Training set:")
print(train_df.head())
print("\nTesting set:")
print(test_df.head())
```

```python
import random
import pandas as pd

class WeightedRandomSplitter:
    def __init__(self, df, target_column, weight_column, train_size=0.7):
        """
        Initialize the splitter with a DataFrame and the target column to use for stratified weighting.

        Parameters:
        df (pd.DataFrame): DataFrame containing the dataset.
        target_column (str): Name of the target column to use for stratification.
        weight_column (str): Name of the column to use for weighting the split.
        train_size (float): Proportion of the dataset to include in the training set.
        """
        self.df = df
        self.target_column = target_column
        self.weight_column = weight_column
        self.train_size = train_size

    def weighted_random_pick(self, class_df):
        """
        Split the data points of a particular class based on weighted probabilities.

        Parameters:
        class_df (pd.DataFrame): DataFrame containing data points of a particular class.

        Returns:
        train_class_df (pd.DataFrame): Training set DataFrame for the particular class.
        test_class_df (pd.DataFrame): Testing set DataFrame for the particular class.
        """
        train_class_list = []
        test_class_list = []
        total_weight = class_df[self.weight_column].sum()

        for _, row in class_df.iterrows():
            weight = row[self.weight_column]
            probability = weight / total_weight

            if len(train_class_list) / len(class_df) < self.train_size and random.choices([True, False], weights=[probability, 1 - probability], k=1)[0]:
                train_class_list.append(row)
            else:
                test_class_list.append(row)

        train_class_df = pd.DataFrame(train_class_list, columns=class_df.columns)
        test_class_df = pd.DataFrame(test_class_list, columns=class_df.columns)

        return train_class_df, test_class_df

    def split_dataset(self):
        """
        Split the dataset into training and testing sets based on stratified weighted probabilities.

        Returns:
        train_df (pd.DataFrame): Training set DataFrame.
        test_df (pd.DataFrame): Testing set DataFrame.
        """
        train_df = pd.DataFrame(columns=self.df.columns)
        test_df = pd.DataFrame(columns=self.df.columns)

        for target_class in self.df[self.target_column].unique():
            class_df = self.df[self.df[self.target_column] == target_class]
            train_class_df, test_class_df = self.weighted_random_pick(class_df)
            train_df = pd.concat([train_df, train_class_df], ignore_index=True)
            test_df = pd.concat([test_df, test_class_df], ignore_index=True)

        return train_df, test_df

# Example usage with a small synthetic dataset
data = {
    'sample': ['S1', 'S2', 'S3', 'S4', 'S5', 'S6'],
    'class': ['A', 'A', 'A', 'B', 'B', 'B'],
    'weight': [10, 5, 1, 8, 2, 1]
}
df = pd.DataFrame(data)

# Instantiate the splitter with the DataFrame, target column, and weight column
splitter = WeightedRandomSplitter(df=df, target_column='class', weight_column='weight', train_size=0.7)

# Split the dataset into training and testing sets
train_df, test_df = splitter.split_dataset()

# Display the results
print("Training set:")
print(train_df)
print("\nTesting set:")
print(test_df)
```

```python
import random
import pandas as pd
from sklearn.datasets import load_iris

class WeightedRandomSplitter:
    def __init__(self, df, target_column, weight_column, train_size=0.7):
        """
        Initialize the splitter with a DataFrame, target column, weight column, and training set size.

        Parameters:
        df (pd.DataFrame): DataFrame containing the dataset.
        target_column (str): Name of the target column to use for stratification.
        weight_column (str): Name of the column to use for weighting the split.
        train_size (float): Proportion of the dataset to include in the training set.
        """
        self.df = df
        self.target_column = target_column
        self.weight_column = weight_column
        self.train_size = train_size

    def split_class(self, class_df):
        """
        Split data points of a particular class based on weighted probabilities.

        Parameters:
        class_df (pd.DataFrame): DataFrame containing data points of a particular class.

        Returns:
        train_class_df (pd.DataFrame): Training set DataFrame for the class.
        test_class_df (pd.DataFrame): Testing set DataFrame for the class.
        """
        total_samples = len(class_df)
        train_samples = int(total_samples * self.train_size)

        # Sort by weight to ensure reproducibility
        class_df = class_df.sample(frac=1, random_state=1).reset_index(drop=True)
        weights = class_df[self.weight_column].values
        total_weight = sum(weights)
        probabilities = [w / total_weight for w in weights]

        train_indices = random.choices(range(total_samples), weights=probabilities, k=train_samples)
        train_indices = sorted(set(train_indices))  # Ensure unique indices
        train_class_df = class_df.loc[train_indices]
        test_class_df = class_df.drop(train_indices).reset_index(drop=True)

        return train_class_df.reset_index(drop=True), test_class_df

    def split_dataset(self):
        """
        Split the dataset into training and testing sets based on stratified weighted probabilities.

        Returns:
        train_df (pd.DataFrame): Training set DataFrame.
        test_df (pd.DataFrame): Testing set DataFrame.
        """
        train_df = pd.DataFrame(columns=self.df.columns)
        test_df = pd.DataFrame(columns=self.df.columns)

        for target_class in self.df[self.target_column].unique():
            class_df = self.df[self.df[self.target_column] == target_class]
            train_class_df, test_class_df = self.split_class(class_df)
            train_df = pd.concat([train_df, train_class_df], ignore_index=True)
            test_df = pd.concat([test_df, test_class_df], ignore_index=True)

        return train_df, test_df

# Example usage with the Iris dataset
# Load the Iris dataset and convert it to a DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Sample weights for the sake of example
df['weight'] = df['sepal length (cm)']

# Instantiate the splitter with the DataFrame and the target column
splitter = WeightedRandomSplitter(df=df, target_column='species', weight_column='weight', train_size=0.7)

# Split the dataset into training and testing sets
train_df, test_df = splitter.split_dataset()

# Display the results
print("Training set:")
print(train_df)
print("\nTesting set:")
print(test_df)
```

```python
# Import the random module for random choices and pandas for data manipulation
import random
import pandas as pd

# Define a class for splitting a dataset based on stratified weighted probabilities
class WeightedStratifiedSplit:
    def __init__(self, target_column, weight_column, df, train_size=0.7):
        # Initialize the class with a DataFrame, target column, weight column, and training set size
        self.df = df
        self.target_column = target_column
        self.weight_column = weight_column
        self.train_size = train_size  # Ensure the parameter train_size is used correctly

    # Method to split data points of a particular class based on weighted probabilities
    def WeightclassSplit(self, class_df):
        # Initialize lists to collect training and testing data points
        train_class_split = []
        test_class_split = []
        # Calculate the total weight for the class
        total_weight = class_df[self.weight_column].sum()
        # Calculate the number of samples to include in the training set
        num_train_samples = int(len(class_df) * self.train_size)
        # print(num_train_samples)
        # print(class_df[self.target_column].unique())
        # Convert the DataFrame to a list of rows to manipulate easily
        class_list = class_df.to_dict('records')

        # Shuffle the list to randomize the selection process
        random.shuffle(class_list)

        while len(train_class_split) < num_train_samples:
            for row in class_list:
                if len(train_class_split) <= num_train_samples:
                    weight = row[self.weight_column]
                    probability = weight / total_weight
                    if random.choices([True, False], weights=[probability, 1 - probability], k=1)[0]:
                        train_class_split.append(row)
                        class_list.remove(row)
                else:
                    break

        # Remaining rows go to the test set
        test_class_split = class_list

        # Convert the lists back to DataFrames
        train_class_df = pd.DataFrame(train_class_split)
        test_class_df = pd.DataFrame(test_class_split)

        # Return the training and testing DataFrames for the class
        return train_class_df, test_class_df

    # Method to split the entire dataset into training and testing sets based on stratified weighted probabilities
    def splitter(self):
        # Initialize empty DataFrames for the training and testing sets
        train_df = pd.DataFrame(columns=self.df.columns)
        test_df = pd.DataFrame(columns=self.df.columns)

        # Iterate over each unique class in the target column
        for cls in self.df[self.target_column].unique():
            # Get the subset of data points for the current class
            class_df = self.df[self.df[self.target_column] == cls]
            # Split the class-specific data into training and testing sets
            train_df_class, test_df_class = self.WeightclassSplit(class_df)
            # Append the training data for the class to the overall training DataFrame
            train_df = pd.concat([train_df, train_df_class], ignore_index=True)
            # Append the testing data for the class to the overall testing DataFrame
            test_df = pd.concat([test_df, test_df_class], ignore_index=True)

        # Return the training and testing DataFrames
        return train_df, test_df

# Example usage with a small synthetic dataset
data = {
    'sample': ['S1', 'S2', 'S3', 'S4', 'S5', 'S6','S7','S8','S9','S10'],
    'class': ['A', 'A', 'A', 'B', 'B', 'B','C','C','C','C'],
    'weight': [10, 5, 1, 8, 2, 1,10,20,30,40]
}
df = pd.DataFrame(data)

# Instantiate the splitter with the DataFrame, target column, and weight column
splitter = WeightedStratifiedSplit(df=df, target_column='class', weight_column='weight', train_size=0.7)  # Ensure train_size is 0.7

# Split the dataset into training and testing sets
train_df, test_df = splitter.splitter()  # Call the splitter method and assign the result to train_df and test_df

# Display the results
print("Training set:")
print(train_df)
print("\nTesting set:")
print(test_df)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
import random

class BootstrapSampler:
    def __init__(self, df, n_samples, sample_size):
        """
        Initialize the BootstrapSampler with a DataFrame, number of samples, and sample size.

        Parameters:
        df (pd.DataFrame): DataFrame containing the dataset.
        n_samples (int): Number of bootstrap samples to generate.
        sample_size (int): Size of each bootstrap sample.
        """
        self.df = df
        self.n_samples = n_samples
        self.sample_size = sample_size
        self.samples = []

    def generate_samples(self):
        """
        Generate bootstrap samples by sampling with replacement.
        """
        for _ in range(self.n_samples):
            sample = self.df.sample(n=self.sample_size, replace=True)
            self.samples.append(sample)

    def get_samples(self):
        """
        Return the generated bootstrap samples.

        Returns:
        list of pd.DataFrame: List containing the bootstrap samples as DataFrames.
        """
        return self.samples

# Load the Iris dataset and convert it to a DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Instantiate the BootstrapSampler with the DataFrame, number of samples, and sample size
bootstrap_sampler = BootstrapSampler(df=df, n_samples=5, sample_size=10)

# Generate the bootstrap samples
bootstrap_sampler.generate_samples()

# Get the generated bootstrap samples
samples = bootstrap_sampler.get_samples()

# Display the results
for i, sample in enumerate(samples):
    print(f"Bootstrap Sample {i+1}:")
    print(sample)
    print("\n")
```

```python
import pandas as pd
from sklearn.datasets import load_iris
import random
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

class BootstrapSampler:
    def __init__(self, df, n_samples, sample_size):
        self.df = df
        self.n_samples = n_samples
        self.sample_size = sample_size
        self.samples = []

    def generate_samples(self):
        self.samples = []
        for _ in range(self.n_samples):
            sample = self.df.sample(n=self.sample_size, replace=True)
            self.samples.append(sample)

    def get_samples(self):
        return self.samples

# Load the Iris dataset and convert it to a DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Instantiate the BootstrapSampler with the DataFrame, number of samples, and sample size
bootstrap_sampler = BootstrapSampler(df=df, n_samples=1000, sample_size=len(df))

# Generate the bootstrap samples
bootstrap_sampler.generate_samples()

# Get the generated bootstrap samples
samples = bootstrap_sampler.get_samples()

# Estimate variability of the mean for 'sepal length (cm)'
means = [sample['sepal length (cm)'].mean() for sample in samples]
mean_of_means = np.mean(means)
std_of_means = np.std(means)

print(f"Mean of means: {mean_of_means}")
print(f"Standard deviation of means: {std_of_means}")

# Assess model performance using bootstrap samples
accuracies = []

for sample in samples:
    train_sample = sample
    test_sample = df.drop(sample.index)  # Out-of-bag samples

    model = LogisticRegression(max_iter=200)
    model.fit(train_sample.drop('species', axis=1), train_sample['species'])
    predictions = model.predict(test_sample.drop('species', axis=1))

    accuracy = accuracy_score(test_sample['species'], predictions)
    accuracies.append(accuracy)

mean_accuracy = np.mean(accuracies)
std_accuracy = np.std(accuracies)

print(f"Mean accuracy: {mean_accuracy}")
print(f"Standard deviation of accuracy: {std_accuracy}")

# Construct 95% confidence interval for the mean of 'sepal length (cm)'
lower_bound = np.percentile(means, 2.5)
upper_bound = np.percentile(means, 97.5)

print(f"95% confidence interval for the mean: ({lower_bound}, {upper_bound})")
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

class SMOTESampler:
    def __init__(self, k_neighbors=5):
        self.k_neighbors = k_neighbors

    def fit_resample(self, X, y):
        unique_classes, class_counts = np.unique(y, return_counts=True)
        minority_class = unique_classes[np.argmin(class_counts)]
        majority_class = unique_classes[np.argmax(class_counts)]

        # Separating the minority and majority class samples
        X_minority = X[y == minority_class]
        X_majority = X[y == majority_class]

        # Calculate the number of synthetic samples needed
        num_synthetic_samples = len(X_majority) - len(X_minority)

        # Find k-neighbors of the minority class samples
        nbrs = NearestNeighbors(n_neighbors=self.k_neighbors + 1).fit(X_minority)
        _, indices = nbrs.kneighbors(X_minority)

        # Generate synthetic samples
        synthetic_samples = []
        for _ in range(num_synthetic_samples):
            i = np.random.randint(0, len(X_minority))
            nn_index = np.random.choice(indices[i][1:])  # exclude the first neighbor (itself)
            diff = X_minority[nn_index] - X_minority[i]
            gap = np.random.rand()
            synthetic_sample = X_minority[i] + gap * diff
            synthetic_samples.append(synthetic_sample)

        X_synthetic = np.array(synthetic_samples)
        y_synthetic = np.array([minority_class] * len(synthetic_samples))

        # Combine original and synthetic samples
        X_resampled = np.vstack((X_majority, X_minority, X_synthetic))
        y_resampled = np.hstack(([majority_class] * len(X_majority), [minority_class] * len(X_minority), y_synthetic))

        return X_resampled, y_resampled

# Create a synthetic, imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, weights=[0.9, 0.1], flip_y=0, random_state=42)

# Check class distribution before SMOTE
print("Class distribution before SMOTE:", np.bincount(y))

# Apply SMOTE
smote_sampler = SMOTESampler(k_neighbors=5)
X_resampled, y_resampled = smote_sampler.fit_resample(X, y)

# Check class distribution after SMOTE
print("Class distribution after SMOTE:", np.bincount(y_resampled))

# Optional: Visualize the original and synthetic samples for the first two PCA components
from sklearn.decomposition import PCA

def plot_2d_space(X, y, label='Classes'):
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    colors = ['r', 'b']
    markers = ['o', 's']
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X_pca[y == l, 0],
            X_pca[y == l, 1],
            c=c, label=f'Class {l}', marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()

plot_2d_space(X, y, 'Original Imbalanced Dataset')
plot_2d_space(X_resampled, y_resampled, 'SMOTE Resampled Dataset')
```

```python
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation and analysis
from sklearn.neighbors import NearestNeighbors  # For nearest neighbors calculation
from sklearn.datasets import make_classification  # For creating a synthetic dataset

# Define the SMOTESampler class for balancing datasets using the SMOTE technique.
class SMOTESampler:
    # Initialize the SMOTESampler class with a specified number of neighbors.
    def __init__(self, n_neighbors):
        self.n_neighbors = n_neighbors  # int

    # Define the fit_resample method to balance the dataset by generating synthetic samples for the minority class.
    def fit_resample(self, X, y):
        # Get unique classes and their counts from the target array.
        unique_classes = y.iloc[:, 0].unique()  # np.ndarray

        # Get counts of each class
        class_counts = y.iloc[:, 0].value_counts().values  # np.ndarray

        # Identify the minority class (the class with the fewest samples).
        minority_class = unique_classes[np.argmin(class_counts)]  # Element from np.ndarray, same dtype as y column

        # Identify the majority class (the class with the most samples).
        majority_class = unique_classes[np.argmax(class_counts)]  # Element from np.ndarray, same dtype as y column

        # Separate the minority class samples from the majority class samples.
        X_minority = X[y.iloc[:, 0] == minority_class]  # pd.DataFrame
        X_majority = X[y.iloc[:, 0] == majority_class]  # pd.DataFrame

        # Check if X_minority or X_majority is empty to avoid errors.
        if X_minority.empty or X_majority.empty:  # pd.DataFrame.empty
            raise ValueError("Found array with 0 sample(s). Check the data and class distribution.")

        # Calculate the number of synthetic samples needed to balance the dataset.
        num_synthetic_samples = len(X_majority) - len(X_minority)  # int

        # Initialize the NearestNeighbors model to find k-neighbors of the minority class samples.
        nbrs = NearestNeighbors(n_neighbors=self.n_neighbors + 1).fit(X_minority)  # NearestNeighbors
        _, indices = nbrs.kneighbors(X_minority)  # _, np.ndarray

        # Generate synthetic samples.
        synthetic_samples = []  # list of np.ndarrays
        while len(synthetic_samples) < num_synthetic_samples:
            i = np.random.randint(0, len(X_minority))  # int
            nn_index = np.random.choice(indices[i][1:])  # int, np.ndarray slicing
            diff = X_minority.iloc[nn_index] - X_minority.iloc[i]  # pd.Series
            gap = np.random.rand()  # float
            synthetic_sample = X_minority.iloc[i] + gap * diff  # pd.Series
            synthetic_samples.append(synthetic_sample)  # Appending pd.Series to list

        # Convert synthetic samples to a DataFrame.
        X_synthetic = pd.DataFrame(synthetic_samples, columns=X.columns)  # pd.DataFrame

        # Create an array for the labels of the synthetic samples.
        y_synthetic = pd.DataFrame([minority_class] * len(X_synthetic), columns=y.columns)  # pd.DataFrame

        # Combine original and synthetic samples.
        X_resampled = pd.concat([X_majority, X_minority, X_synthetic])  # pd.DataFrame
        y_resampled = pd.concat([y[y.iloc[:, 0] == majority_class], y[y.iloc[:, 0] == minority_class], y_synthetic])  # pd.DataFrame

        return X_resampled, y_resampled  # tuple of pd.DataFrames

# Create a synthetic, imbalanced dataset.
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, weights=[0.9, 0.1], flip_y=0, random_state=42)

# Convert to DataFrame.
df_X = pd.DataFrame(X)  # pd.DataFrame
df_y = pd.DataFrame(y, columns=['Class'])  # pd.DataFrame

# Check class distribution before SMOTE.
print("Class distribution before SMOTE:\n", df_y['Class'].value_counts())  # pd.Series

# Apply SMOTE.
smote_sampler = SMOTESampler(n_neighbors=5)  # SMOTESampler instance
X_resampled, y_resampled = smote_sampler.fit_resample(df_X, df_y)  # tuple of pd.DataFrames

# Check class distribution after SMOTE.
print("Class distribution after SMOTE:\n", y_resampled['Class'].value_counts())  # pd.Series
```
